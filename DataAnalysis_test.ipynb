Import DATABASE: (vendas.xlsx)

```python
import pandas as pd

table = pd.read_excel("DB/vendas.xlsx")
```

Get overview of information.

```python
total_billing = table["Valor Final"].sum()
total_sales = table["Quantidade"].sum()

print('TOTAL R$: '+str(total_billing))
print("TOTAL Amount: "+str(total_sales))
```

Analyze Top-> Down

```python
#billing per store and items amount
billing_per_store = table[["ID Loja","Valor Final","Quantidade"]].groupby("ID Loja").sum("Valor Final")
display(billing_per_store)
```

```python
#Enter the details
billing_product = table[["ID Loja","Produto","Valor Final"]].groupby(["ID Loja", "Produto"]).sum()
display(billing_product)
```

Graphics

```python
products = table[["Produto","Quantidade"]].groupby("Produto").sum("Quantidade")
df = pd.DataFrame(products)
```

Best selling items

```python
df.plot.pie(y='Quantidade',figsize=(20, 15),autopct="%.2f")
```
